```python
from apyori import apriori
```

```python
store_data = pd.read_csv('store_data.csv', header=None)
```

```python
store_data.head()
```

```python
records = []
for i in range(0, 7501):
    records.append([str(store_data.values[i,j]) for j in range(0, 20)])
```

```python
# Setting up min support, confidence and lift to pull only rules that meet the minimum criteria
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
#Checking the total number of association rules pulled by using the above settings for support, confidence and lift
print(len(association_results))
```

```python
#Printing all the 48 association rules
print(association_results)
```

```python
#Printing the 1st association rule
print(association_results[0])
```

```python
#Printing the 2nd association rule
print(association_results[1])
```

```python
#Pulling together all association rules
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    #Support can be calculated by finding number of transactions containing particular item divided by total no of transactions
    #Support(product A)= (Transactions containing (product A)/(Total transactions))
    #In below example - Support (light cream-->chicken)= (Transactions containing cream-chese)/Total number of transactions
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    # Confidence(A-->B) refers to likelihood of that an item B is also bought if item A is bought
    print("Confidence: " + str(item[2][0][2]))
    
    # Lift(A-->B) refers to the increase in the ratio of sale of B when A is sold.
    # Lift can be calculated by dividing Confidence (A-->B) divided by Support (B)
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
